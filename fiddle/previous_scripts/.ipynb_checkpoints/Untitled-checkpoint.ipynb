```python
import numpy as np
import pandas as pd
import pybedtools as pb
```

```python
pb.filenames.example_filename('c.gff')
```

```python
annot = pd.read_csv('/Users/umut/Projects/CellCycle/data/AllTranscriptsSteinmetz.csv',',')
```

```python
annot.sort_values(by =['chr','start'],inplace=True)
```

```python
annot.head()
```

```python
df = pd.DataFrame()
stride = 20
window_size = 500
for ch in np.arange(1,17):
    mn = np.min(annot.loc[annot['chr']==ch,'start'])
    mx = np.max(annot.loc[annot['chr']==ch,'end'])
    starts = np.arange(mn,mx,stride)
    df = df.append(pd.DataFrame({'chr':[ch]*len(starts),'start':starts,'end':starts+window_size}))
```
